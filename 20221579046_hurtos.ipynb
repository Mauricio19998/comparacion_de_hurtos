1. Realizar la comparación de hurtos entre Medellín, Bogotá y Cali para el año 2020 y sacar tres conclusiones. 

```python
# Se cambia el nombre de la columna en los datos para que no nos salga error
hurtos_nuevo = hurtos.rename(columns={
    "FECHA HECHO": "Fecha"
})
hurtos_nuevo
```

```python
# Se filtran los datos correspondientes a Medellin usando el comnado de query
hurtos_2020_MEDELLIN = hurtos_nuevo.query("Fecha >= '2020-01-01'and Fecha <= '2020-12-31' and MUNICIPIO == 'MEDELLIN'").sort_values("Fecha")
hurtos_2020_MEDELLIN
```

```python
# Se filtran los datos correspondientes a Cali usando el comnado de query
hurtos_2020_CALI = hurtos_nuevo.query("Fecha >= '2020-01-01'and Fecha <= '2020-12-31' and MUNICIPIO == 'CALI'").sort_values("Fecha")
hurtos_2020_CALI
```

```python
# Se filtran los datos correspondientes a Bogota usando el comnado de query
hurtos_2020_bogota = hurtos_nuevo.query("Fecha >= '2020-01-01' and Fecha <= '2020-12-31' and MUNICIPIO == 'BOGOTA, D.C.'").sort_values("Fecha")
hurtos_2020_bogota
```

```python
# Se realiza un grafico comparativo de los datos
plt.plot(hurtos_2020_MEDELLIN["Fecha"], hurtos_2020_MEDELLIN["CANTIDAD"], label="Medellin")# se ingresan los datos de Medellin
plt.plot(hurtos_2020_CALI["Fecha"], hurtos_2020_CALI["CANTIDAD"], label="Cali") # se ingresan los datos de Cali
plt.plot(hurtos_2020_bogota["Fecha"], hurtos_2020_bogota["CANTIDAD"], label="Bogota D.C.") # se ingresan los datos de Bogota
plt.title("Cantidad de hurtos en Medellin, Cali y Bogota año 2020")# Se ingresa el titulo de la grafica
plt.xlabel("Fecha")#se ingresa el nombre de el eje x del grafico
plt.ylabel("Cantidad de hurtos")#se ingresa el nombre de el eje y del grafico
plt.legend()
plt.show()
```

```python
#se reañiza la grafica para poder visualizar los datos de hurtos de una manera mas agradable
total_medellin = hurtos_2020_MEDELLIN['CANTIDAD'].sum())# se ingresan los datos de Medellin
total_cali = hurtos_2020_CALI['CANTIDAD'].sum())# se ingresan los datos de Cali
total_bogota = hurtos_2020_bogota['CANTIDAD'].sum())# se ingresan los datos de Bogota
totales = [total_medellin, total_cali, total_bogota]#se definen los datros que tendra el diagrama
etiquetas = ['Medellin', 'Cali', 'Bogota D.C.']#se ingresan las etiquetas correspondintes
plt.pie(totales, labels=etiquetas, autopct='%1.1f%%')#se maneja la infromacion en porcentajes
plt.title('Distribución de hurtos por ciudad en 2020')#se crea el titulo de la grafica
plt.show()
```

Ingresarlo como diccionario y realizar la división a través de Pandas entre cantidad de hurtos por departamento y población. ¿Qué da como resultado esto? Realizar el diagrama que usted considere más conveniente para mostrar los resultados.

```python
#se importa la libreria csv0
import csv
poblacion_dict = {}
with open('/content/drive/MyDrive/TEORIA Y LOGICA DE PROGRAMACION GRUPO-16/poblacion.csv', 'r', encoding='latin1') as f:
    reader = csv.reader(f)
    next(reader) # saltar la primera fila (encabezados)
    for row in reader:
        poblacion_dict[row[0]] = int(row[1].replace(' ', ''))

print(poblacion_dict)
```

```python
#se crea el dataframe de poblacion
poblacion=pd.DataFrame.from_dict(poblacion, orient="index", columns=["POBLACIÓN"])
poblacion
```

```python
#se crea una lista de la cantidad de hurtos
cant_hurtos=hurtos["DEPARTAMENTO"].value_counts() 
cant_hurtos
```

```python
# se concatena las 2 listas
hurtos_poblacion=pd.concat([cant_hurtos, poblacion], axis=1)
hurtos_poblacion
```

```python
#se usa el comando de division para poder hacer la compracion de huirtos por cada 10mil habitantes
hurtos["Division"]=(hurtos_poblacion["DEPARTAMENTO"]/hurtos_poblacion["POBLACIÓN"])*10000
hurtos
```
